Exploratory Data Analysis (EDA)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("customers.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
print(data.duplicated())  # Count duplicate rows
```

```python
print(data.duplicated().sum())
```

```python
print(data.describe())  # Summary for numerical columns
```

```python
print(data.describe(include="object"))  # Summary for categorical columns
```

```python
salary = pd.Series(np.random.uniform(1000, 10000, 10000), name="Salary")
```

```python
data=pd.concat([data, salary], axis=1)
```

```python
data
```

```python
data["Salary"]
```

```python
data.describe()
```

```python
data["Salary"][8000] = 30000.0

sns.boxplot(x=data["Salary"])
plt.show()
```

```python
Q1 = data["Salary"].quantile(0.25)
Q3 = data["Salary"].quantile(0.75)
IQR = Q3 - Q1
df = data[(data["Salary"] >= Q1 - 1.5 * IQR) & (data["Salary"] <= Q3 + 1.5 * IQR)]
```

```python
sns.boxplot(x=df["Salary"])
plt.show()
```

```python
df.hist(figsize=(10, 6), bins=30)
plt.show()
```
